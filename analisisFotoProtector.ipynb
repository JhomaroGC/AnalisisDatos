```python
import pandas as pd
from datetime import datetime
```

```python
file = "content\data.csv"
df = pd.read_csv(file)
df['fecha'] = df['Created'].str.split(' ').str[0]
# df['fecha'] = pd.to_datetime(df['fecha'])
df['nombre'] = df['solicitante']
df = df[df['Producto'] == 'FOTOPROTECTOR 50 ISDIN']
df = df[df['estado'] == 'Aprobado']
df = df[df['entregado'] == 'Sí']
df = df[['fecha', 'nombre', 'cantidad']]
df = df.sort_values(by='cantidad', ascending=False)
df.head()
```

```python
df_agrupado = df.groupby(['nombre'])['fecha'].apply(list).reset_index()
df_agrupado['Cantidad'] = df_agrupado['fecha'].apply(len)
df_agrupado = df_agrupado.sort_values(by='Cantidad', ascending=False)
df_agrupado.head()
```

```python


df_agrupado.to_csv(f"content/fotoProtector{datetime.now().strftime('%Y%m%d%H%M%S')}.csv", index=False)
```
